```python
def clean_text(text):
    cleaned_text = re.sub(' ', '', text)
    cleaned_text = re.sub('\-\#', '', cleaned_text)
    cleaned_text = re.sub('[\{\}\[\]\/?,;:|\)*~`!^_+<>@$%\\\=\(\'\"]',
                          ' ', cleaned_text)
    return cleaned_text.strip()
```
