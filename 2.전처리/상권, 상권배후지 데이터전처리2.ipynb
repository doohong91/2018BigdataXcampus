```python
import pandas as pd
```

```python
data=[[]for i in range(0,14)]
data_temp=[[]for i in range(0,14)]
```

```python
for i in range(0,7):
    data[i]=pd.read_excel("../data/data/" +str(i+1)+".xlsx")
    data[i+7]=pd.read_excel("../data/data/" +str(i+1)+str(i+1)+".xlsx")
```

```python
for i in range(0,14):
    data[i]=data[i].drop(data[i][data[i]['TRDAR_CD_NM'].isnull()].index)
```

```python
for i in range(0,14):
    data_temp[i]=data[i]
    data_temp[i]['인덱스']=data[i][data[i].columns[0]].astype(str).str.cat(data[i][data[i].columns[1]].astype(str))
    data_temp[i] =data_temp[i].set_index(['인덱스'])
    data_temp[i] =data_temp[i].groupby(level=0).last()
```

```python
data_temp[6]=data[6]
data_temp[6]['임시 인덱스']=data[6][data[6].columns[0]].astype(str).str.cat(data[6][data[6].columns[1]].astype(str).str.cat(data[6][data[6].columns[2]]))
data_temp[6]['인덱스']=data[6][data[6].columns[0]].astype(str).str.cat(data[6][data[6].columns[1]].astype(str))
data_temp[6] = data_temp[6].set_index(['임시 인덱스'])
data_temp[6] =data_temp[6].groupby(level=0).last()
data_temp[13]=data[13]
data_temp[13]['임시 인덱스']=data[13][data[13].columns[0]].astype(str).str.cat(data[13][data[13].columns[1]].astype(str).str.cat(data[13][data[13].columns[2]]))
data_temp[13]['인덱스']=data[13][data[13].columns[0]].astype(str).str.cat(data[13][data[13].columns[1]].astype(str))
data_temp[13]= data_temp[13].set_index(['임시 인덱스'])
data_temp[13] =data_temp[13].groupby(level=0).last()
```

```python
del_list =['STDR_YM_CD','TRDAR_CD_NM']

for i in range(0,6):
    for j in del_list:
        del data_temp[i][j]
        del data_temp[i+7][j]
```

```python
data1 = pd.concat([data_temp[0],data_temp[1],data_temp[2],data_temp[3],data_temp[4],data_temp[5]],axis=1,sort=False,join='inner')
```

```python
data11 = pd.concat([data_temp[7],data_temp[8],data_temp[9],data_temp[10],data_temp[11],data_temp[12]],axis=1,sort=False, join='inner')
```

```python
data2 = data1
data22 = data11
```

```python
del_list =['STDR_YM_CD','TRDAR_CD_NM']

for j in del_list:
    del data_temp[6][j]
    del data_temp[13][j]
```

```python
data3 = pd.pivot_table(data_temp[6],index=['인덱스'],columns=['SVC_INDUTY_CD_NM'])
```

```python
data33 = pd.pivot_table(data_temp[13],index=['인덱스'],columns=['SVC_INDUTY_CD_NM'])
```

```python
data_final1 = pd.concat([data2,data3],axis=1, join='inner')
```

```python
data_final1.to_excel("../data/data/data_final.xlsx")
```

```python
data_final2 = pd.concat([data22,data33],axis=1, join='inner')
```
