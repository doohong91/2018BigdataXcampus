```python
brand1.columns = ['상호', '자산','매출액', '영업이익', '총자산증가율', '매출액증가율', '영업이익증가율','중분류']
brand2.columns = ['상호_2','자본','매출액_2','영업이익_2','당기순이익','영업이익률(영업이익/매출액)','매출액순이익률(당기순이익/매출액)','자기자본순이익률(당기순이익/자본)','중분류_2']
brand3.columns = ['상호_3','자산_2','자본_2','부채','부채비율(부채/자본)','자기자본비율(자본/자산)','법위반횟수(최근3년)','중분류_3']
```

```python
brand1.shape, brand2.shape, brand3.shape
```

```python
brand12 = pd.merge(brand1,brand2, left_on=["중분류",'상호'], right_on=["중분류_2",'상호_2'], how="left")
brand123 = pd.merge(brand12,brand3, left_on=["중분류",'상호'], right_on=["중분류_3",'상호_3'], how="left")
del brand123['상호_2']
del brand123['상호_3']
del brand123['중분류_2']
del brand123['중분류_3']
del brand123['자본_2']
del brand123['자산_2']
del brand123['매출액_2']
del brand123['영업이익_2']
# #del b2016['중분류_2']
# #del b2016['중분류_3']
brand123.shape
```
