```python
import numpy as np
import pandas as pd
```

```python
brand1 = pd.read_csv('../data/brand/2017/2017_1.csv',index_col=0) 
brand2 = pd.read_csv('../data/brand/2017/2017_2.csv',index_col=0)
brand3 = pd.read_csv('../data/brand/2017/2017_3.csv',index_col=0) 
```

```python
brand1.drop(brand1[brand1['0'] == '평균'].index, inplace=True)
```

```python
brand1.head()
```

```python
brand2.drop(brand2[brand2['0'] == '평균'].index, inplace=True)
```

```python
brand2.head()
```

```python
brand3.drop(brand3[brand3['0'] == '평균'].index, inplace=True)
```

```python
brand3.head()
```

```python
brand1.columns = ['브랜드','상호','가맹사업 개시일','가맹사업 년수','가맹점수','가맹본부 임직원수']
brand2.columns = ['브랜드_2','상호_2','가맹점수_2','신규개점','계약종료','계약해지','명의변경','가맹점 평균매출액(매출액지수)','가맹점 면적(3.3㎡)당 평균매출액(평균 매출액지수)']
brand3.columns = ['브랜드_3','상호_3','가입비(가맹비)','교육비','보증금', '기타비용(인테리어 비용포함)', '합계 (창업비용 지수)', '인테리어비용 면적당(3.3㎡) 비용', '인테리어비용 기준면적 (㎡)', '인테리어비용 총 비용','중분류']
```

```python
brand1.shape, brand2.shape, brand3.shape
```

```python
brand12 = pd.merge(brand1,brand2, left_on=["브랜드",'상호'], right_on=["브랜드_2",'상호_2'], how="left")
brand123 = pd.merge(brand12,brand3, left_on=["브랜드",'상호'], right_on=["브랜드_3",'상호_3'], how="left")
del brand123['상호_2']
del brand123['상호_3']
del brand123['브랜드_2']
del brand123['브랜드_3']
del brand123['가맹점수_2']
# #del b2016['중분류_2']
# #del b2016['중분류_3']
brand123.shape
```
