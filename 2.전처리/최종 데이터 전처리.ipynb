```python
data = pd.read_excel('./최종데이터.xlsx')
```

```python
data['공정위시정조치'].fillna(0, inplace = True)as
```

```python
data['민사소송패소/민사상화홰'].fillna(0, inplace = True)
```

```python
data['형의 선고'].fillna(0, inplace = True)
```

```python
pd.get_dummies(data.중분류).head()
```

```python
data = pd.concat([data,pd.get_dummies(data.중분류)],axis=1)
```

```python
data[np.isnan(data['자산'])]
```

```python
data['법위반횟수(최근3년)'] = data['공정위시정조치']+data['민사소송패소/민사상화홰']+data['형의 선고']
```

```python
data['창업비용 지수'] = data['비용 합계']/np.mean(data['비용 합계'])
```

```python
data['매출액지수'] = data['가맹점 평균매출액']/np.mean(data['가맹점 평균매출액'])
```

```python
data['영업이익률(영업이익/매출액)'] = data['영업이익']/data['매출액']
```

```python
data['매출액순이익률(당기순이익/매출액)'] = data['당기순이익']/data['매출액']
```

```python
data['자기자본순이익률(당기순이익/자본)'] = data['당기순이익']/data['자본']
```

```python
data['부채비율(부채/자본)'] = data['부채']/data['자본']
```

```python
data['자기자본비율(자본/자산)'] = data['자본']/data['자산']
```

```python
data.to_excel('recaldata.xlsx')
```
