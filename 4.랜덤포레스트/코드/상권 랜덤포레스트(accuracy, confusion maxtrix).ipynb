```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import scale, robust_scale, minmax_scale, maxabs_scale
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
X = pd.read_excel("../301-8/data/data/4단계X.xlsx",index_col='인덱스')
Y = pd.read_excel("../301-8/data/data/4단계Y.xlsx",index_col='인덱스')
```

```python
X
```

```python
Y
```

```python
Report = pd.DataFrame(index=X.columns)
Score  = pd.DataFrame(index=['훈련 acc','검증 acc'])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=0)
```

```python
for i in range(0,len(Y.columns)):
    clf = RandomForestClassifier(n_estimators = 30, random_state=0)
    clf.fit(X_train, Y_train[Y_train.columns[i]])
    Report[Y_train.columns[i]]=clf.feature_importances_
    Score[Y_train.columns[i]]=[clf.score(X_train, Y_train[Y_train.columns[i]]),clf.score(X_test, Y_test[Y_test.columns[i]])]
```

```python
Score.to_excel("../301-8//data/data/acc(랜덤레스트).xlsx")
Score
```

```python
Report.to_excel("../301-8//data/data/importances(랜덤포레스트).xlsx")
Report
```

```python
clf = RandomForestClassifier(n_estimators = 30, random_state=0)
clf.fit(X_train, Y_train[Y_train.columns[6]])
print(confusion_matrix(Y_test[Y_test.columns[6]], clf.predict(X_test)))
print(classification_report(Y_test[Y_test.columns[6]], clf.predict(X_test)))
```
