```python
driver = webdriver.Chrome('../../../driver/chromedriver')
driver.get("http://franchise.ftc.go.kr/user/extra/main/70/firHope/listBrand/jsp/LayOutPage.do")
```

```python
year_list_raw = driver.find_element_by_xpath("""//*[@id="selYear"]""")
year_list = year_list_raw.find_elements_by_tag_name("option")
year_names_values = [option.text for option in year_list]
year_names_values = year_names_values[:]
year_names_values
```

```python
indus_list_raw = driver.find_element_by_xpath("""//*[@id="selIndus"]""")
indus_list = indus_list_raw.find_elements_by_tag_name("option")
indus_names_values = [option.text for option in indus_list]
indus_names_values = indus_names_values[1:]
indus_names_values
```

```python
report_list_raw = driver.find_element_by_xpath("""//*[@id="selListType"]""")
report_list = report_list_raw.find_elements_by_tag_name("option")
report_names_values = [option.text for option in report_list]
report_names_values = report_names_values[:]
report_names_values
```

```python
r = []
length = [6, 9, 10]
k=2
year = driver.find_element_by_id("selYear") #년도
year.send_keys(year_names_values[0])
    
for j in range(0,len(indus_names_values)):
    indus = driver.find_element_by_id("selIndus") #업종 중분류
    indus.send_keys(indus_names_values[j])
        
#    for k in range(0,len(report_names_values)):
    report = driver.find_element_by_id("selListType") #비교항목
    report.send_keys(report_names_values[2])
    driver.find_element_by_xpath("""//*[@id="frmSearch"]/input[5]""").click()
            
    html=driver.page_source
    soup=BeautifulSoup(html,'html.parser')
            
    print(report_names_values[2])####

    rows = []        
    for row in range(int(len(soup.find_all('td'))/length[k])):
        row1 = []

        for i in range(0+(row*length[k]), length[k]+(row*length[k])) :
            tmp = soup.find_all('td')[i].get_text()
            tmp = tmp.replace('\n','')
            tmp = tmp.replace('\t','')
            tmp = tmp.replace('\xa0','')

            row1.append(tmp)
        rows.append(row1)
    r.append(rows)
    print(indus_names_values[j], 'done!')
```
