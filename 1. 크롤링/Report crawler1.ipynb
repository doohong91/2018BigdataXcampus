```python
r = []
r1 = []
r2 = [] 
r3 = []
r4 = []
r5 = []
for url in list1[0:10]:
    html = urlopen(url)
    soup = BeautifulSoup(html,'html.parser')
    
    row=[]
    #가맹본부 일반 현황
    row1=[]
    for i in range(0,11) :
        tmp = soup.find_all('td')[i].get_text()
        tmp1 = tmp.split('\r')[1]    
        tmp1 = tmp1.replace('\n','')
        tmp1 = tmp1.replace('\t','')
        tmp1 = tmp1.replace('\xa0','')
        if tmp1 == '' :
            tmp1 = tmp.split('\r')[2]
            tmp1 = tmp1.replace('\n','')
            tmp1 = tmp1.replace('\t','')     
        row1.append(tmp1)
        row.append(tmp1)
    #주소, 사업자유형, 법인등록번호, 사업자등록번호
    row2=[]
    for i in range(12,16) :
        tmp = soup.find_all('td')[i].get_text()
        tmp1 = tmp.split('\r')[1]    
        tmp1 = tmp1.replace('\n','')
        tmp1 = tmp1.replace('\t','')
        tmp1 = tmp1.replace('\xa0','')
        if tmp1 == '' :
            tmp1 = tmp.split('\r')[2]
            tmp1 = tmp1.replace('\n','')
            tmp1 = tmp1.replace('\t','')
            tmp1 = tmp1.replace('\xa0','')
        row2.append(tmp1)
        row.append(tmp1)
    #가맹본부 재무상황
    row3=[]
    for i in range(16,40) :
        tmp = soup.find_all('td')[i].get_text()
        if tmp == '' :           
            row3.append(tmp)
            row.append(tmp)
            continue
        try :
            tmp1 = tmp.split('\r')[1]    
        except : 
            tmp1 = tmp
        tmp1 = tmp1.replace('\n','')
        tmp1 = tmp1.replace('\t','')
        if tmp1 == '' :
            tmp1 = tmp.split('\r')[2]
            tmp1 = tmp1.replace('\n','')
            tmp1 = tmp1.replace('\t','')    
        row3.append(tmp1)
        row.append(tmp1)
    #가맹사업 임직원수, 가맹본부 브랜드 및 가맹사업 계열사 수, 가맹사업 개시일
    row4=[]
    for i in range(40,46) :
        tmp = soup.find_all('td')[i].get_text()
        if (tmp == '')|(tmp=='\n') :
            row4.append('')
            row.append('')
            continue
        try :
            tmp1 = tmp.split('\r')[1]    
        except : 
            tmp1 = tmp
        tmp1 = tmp1.replace('\n','')
        tmp1 = tmp1.replace('\t','')
        if tmp1 == '' :
            tmp1 = tmp.split('\r')[2]
            tmp1 = tmp1.replace('\n','')
            tmp1 = tmp1.replace('\t','')    
        row4.append(tmp1)
        row.append(tmp1)
        
    #그외
    row5=[]
    for i in range(46,len(soup.find_all('td'))) :
        tmp = soup.find_all('td')[i].get_text()
        tmp = tmp.replace('\n','')
        tmp = tmp.replace('\t','')
        tmp = tmp.replace('\r','')
        tmp = tmp.replace('\xa0','')    
        row5.append(tmp)
        row.append(tmp)
    
    r1.append(row1)
    r2.append(row2)
    r3.append(row3)
    r4.append(row4)
    r5.append(row5)
    r.append(row)
```

```python
data_frame = pd.DataFrame(r)
data_frame.to_csv('가나다.csv',encoding='utf-8')
```
